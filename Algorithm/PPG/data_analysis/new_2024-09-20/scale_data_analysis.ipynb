```python
import json
import copy
import pandas as pd
from datetime import datetime
```

```python
# 读取导出的 JSON 文件
directory = r'C:\Users\Administrator\Desktop\middle_school_data\遂昌民族中学.json'
with open(directory, 'r', encoding='utf-8') as file:
    data = json.load(file)
```

```python
school_name = data["schoolName"]
userAnswers = data["userAnswers"]
```

```python
print("school name: ", school_name)
```

```python
scale_dict = dict()

watch_dict_all = []

for entry in userAnswers:
    userId = entry["userId"]
    userName = entry["userName"]
    birth = datetime.utcfromtimestamp(entry["birth"] / 1000).strftime('%Y-%m-%d')
    gender = entry['gender']
    orgLevels = "/".join(entry["orgLevels"])
    if orgLevels == "遂昌民族中学/测试/测试":
        continue
    print(orgLevels)

    user_profile = {
            "user_id": userId,
            "user_name": userName,
            "birth": birth,
            "class": orgLevels,
            "gender": gender
        }

    scaleAnswers = entry["scaleAnswers"]
    for scale_entry in scaleAnswers:
        scaleName = scale_entry["scaleName"]
        scaleScores = scale_entry["scaleScores"]
        scaleAnswers = scale_entry["scaleAnswers"]        
        scale_score_entries = []
        for score_entry in scaleScores:
            dimensionName = score_entry["dimensionName"]
            score = score_entry["score"]
            ruleName = score_entry["ruleName"]
            scale_score = {
                "dimensionName" : dimensionName, 
                "score": score, 
                "ruleName": ruleName
            }
            scale_score_entries.append(scale_score)
    
        question_answer_entries = []
        for question_entry in scaleAnswers:
            question_title = question_entry["question"]["title"]
            question_index = question_entry["question"]["index"]
            is_reliable_question = question_entry["question"]["isReliableQuestion"]
            options = question_entry["question"]["options"]

            scaleAnswer_answerContent = question_entry["scaleAnswer"]["answerContent"]
            answer_scores = [opt["score"] for opt in options if opt["content"]==scaleAnswer_answerContent]
            assert(len(answer_scores) == 1)
            
            question_answer = {
                "question_title": question_title,
                "question_index": question_index,
                "scaleAnswer_answerContent": scaleAnswer_answerContent,
                "scaleAnswer_answerScore": answer_scores[0], 
                "is_reliable_question": is_reliable_question
            }

            question_answer_entries.append(question_answer)

        if scaleName not in scale_dict.keys():
            scale_dict[scaleName] = list()
        scale_dict[scaleName].append([user_profile, question_answer_entries, scale_score_entries])

    watchResultInfos = entry["watchResultInfos"]
    watch_result_entries = []
    for watch_entry in watchResultInfos:
        startTime = watch_entry["startTime"]
        endTime = watch_entry["endTime"]
        dataResultId = watch_entry["dataResultId"]

        watch_result = {
            "startTime": startTime,
            "endTime": endTime,
            "dataResultId" : dataResultId
        }
        
        watch_result_entries.append(watch_result)

    watch_dict = {
        "user_id": userId,
        "user_name": userName,
        "birth": birth,
        "class": orgLevels, 
        "watch_result_Infos": watch_result_entries
    }
    watch_dict_all.append(watch_dict)
            
```

```python

# 将DataFrame写入Excel文件
with pd.ExcelWriter(r'./outdir/scale_result.xlsx', engine='openpyxl') as writer:    
    for scale_name, scale_entry in scale_dict.items():
        # if scale_name != "抑郁症筛查量表PHQ-9":
        #     continue

        user_dict_all = []
        print("scale name:", scale_name)
        for user_entry in scale_entry:
            user_dict = copy.deepcopy(user_entry[0])   # The user profile
            question_answer_entries = user_entry[1]
            scale_score_entries = user_entry[2]
            
            for answer_entry in question_answer_entries:
                # print(answer_entry)
                question_title = answer_entry["question_title"]
                question_index = answer_entry["question_index"]
                scaleAnswer_answerContent = answer_entry["scaleAnswer_answerContent"]
                scaleAnswer_answerScore = answer_entry["scaleAnswer_answerScore"]

                is_reliable_question = answer_entry["is_reliable_question"]
                
                user_dict[question_title] = scaleAnswer_answerContent
                user_dict["question_" + str(question_index) + "_reliable"] = is_reliable_question
                user_dict["question_" + str(question_index) + "_answer_core"] = scaleAnswer_answerScore
    
            for score_entry in scale_score_entries:
                dimensionName = score_entry["dimensionName"]
                score = score_entry["score"]
                ruleName = score_entry["ruleName"]
                user_dict[dimensionName] = score
                user_dict["ruleName"] = ruleName
    
            user_dict_all.append(user_dict)
    
        df = pd.DataFrame(user_dict_all)
        df.to_excel(writer, sheet_name=scale_name, index=False)

    df = pd.DataFrame(watch_dict_all)
    df.to_excel(writer, sheet_name="watch_result", index=False)
```
