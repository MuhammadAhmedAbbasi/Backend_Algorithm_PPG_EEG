```python
import json
import os
import pandas as pd

from openpyxl import Workbook
```

```python
file1 = r'C:\Users\Administrator\Desktop\school_data\湖山乡小学.json'
```

```python
import enum

def Convert2xlsx(filePath, fileName):
    wb = Workbook()
    ws = wb.active
    ws.title = "遂昌"
    with open(filePath, 'r', encoding='utf-8-sig') as file:
        datas = json.load(file)['data']['userScaleAnsDataResponses']
    # print(datas)
    columns = ["user_id", 'user_name', 'gender']
    
    user_row_collection = []

    # 遍历每个用户
    for user_idx, data in enumerate(datas):
        # 写入数据
        user_id = data["userId"]
        user_name = data["userName"]
        gender = data["gender"]

        scaleAns = data['scaleAns']

        columns = ["user_id", 'user_name', 'gender']


        dim_score_collection = []

        # 遍历每个用户中的每个量表
        for scale_idx, scaleAns_item in enumerate(scaleAns):
            scaleName = scaleAns_item['scaleName']
            dimensionResults = scaleAns_item['dimensionResults']
            print(scaleName)

            # 1）对于MHT, 查看维度信息
            if scaleName == '心理健康诊断测验-MHT（9-18岁）':
                # 遍历量表中的维度信息
                for dim_idx, dim in enumerate(dimensionResults):
                    
                    dim_score = 0
                    dimension_name = dim['dimensionName']   
                    
                    # print("-------------", dimension_name)
                    questions = dim['questions']
                    for question_items in questions:
                        question = question_items['question']
                        # print(question)
                        title = question['title']
                        isReliableQuestion = question['isReliableQuestion']
                        
                        selectedOption = question_items['selectedOption']
                        selectedOption_score = selectedOption['score']

                        if not isReliableQuestion:
                            dim_score += selectedOption_score
                    dim_score_collection.append(dim_score)

                    if user_idx == 0 and scale_idx == 0:
                        columns.append(dimension_name)

            # 2）对于其他两个焦虑和抑郁量表，计算回答的总分数
            if scaleName == "Spence儿童焦虑量表-简版" or scaleName == "儿童抑郁障碍自评量表":
                if user_idx == 0:
                    columns.append(scaleName)
                questionResults = scaleAns_item['questionResults']
                
                dim_score = 0.
                for question_item in questionResults:
                    isReliableQuestion = question_item['question']['isReliableQuestion']
                    selectedOption_score = question_item['selectedOption']['score']
                    if not isReliableQuestion:
                        dim_score += selectedOption_score
                dim_score_collection.append(dim_score)

        print(columns)       
        user_row = [user_id, user_name, gender] + dim_score_collection
        
        if user_idx == 0:
            ws.append(columns)

        print(user_row)
        ws.append(user_row)

    # 保存为 xlsx 文件
    wb.save(f"{fileName}.xlsx")
    print(f"数据已成功保存到: {fileName}.xlsx 文件中!")


outdir = "outdir"
if not os.path.exists(outdir):
    os.makedirs(outdir)

Convert2xlsx(file1, f"./{outdir}/湖山小学量表结果")
```
